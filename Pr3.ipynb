```python
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook , Workbook
from sklearn.linear_model import LinearRegression
```

```python
ws=load_workbook("Files\linear_regression_data.xlsx")["Sheet1"]
max_row=ws.max_row
print(max_row)
```

```python
data=[]
for i in range( 2 , max_row+1):
    data.append([ws[f'A{i}'].value,ws[f'B{i}'].value,ws[f'C{i}'].value,ws[f'D{i}'].value])
print(data)
```

```python
data_arr=np.array(data)
print(data_arr)
```

```python
inp=data_arr[:,0:3]
output=data_arr[:,3].reshape(10,-1)
print(inp.shape)
print(output.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(inp,output)
```

```python
plt.plot(inp[:,0],output,"o")
plt.title(" input 1 ")
```

```python
plt.plot(inp[:,1],output,"o")
plt.title(" input 2 ")
```

```python
plt.plot(inp[:,2],output,"o")
plt.title(" input 3 ")
```

```python
output_prd=model.predict(inp)
plt.plot(output , output_prd , 'o')
plt.xlabel("output")
plt.ylabel("predicted output")
```

```python
plt.plot(inp[:,0],output,"*")
plt.plot(inp[:,0],output_prd,"o")
plt.title(" input 1 ")
```

```python
plt.plot(inp[:,1],output,"*")
plt.plot(inp[:,1],output_prd,"o")
plt.title(" input 2 ")
```
